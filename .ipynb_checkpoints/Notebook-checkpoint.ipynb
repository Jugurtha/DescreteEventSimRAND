# Générateurs de nombres aléatoires et Simulation de files d'attentes

Dans ce qui suit, nous montrer quelque uns des différents générateurs de nombres aléatoires fournis par le module SciPy de Python, ensuite, nous étudirons ces capacités de simulation en implémentant un système M/M/3 en utilisant le module Simpy.

## Générateurs de nombres pseudo-aléatoires

### Générateur de la loi Uniforme
#### Histogramme :

```python
%matplotlib inline
import numpy as np
import scipy as sp
import scipy.stats as ss
import matplotlib.pyplot as plt

a = 5
b = 10
uniform = ss.uniform(loc = a, scale =  b - a)
plt.hist(uniform.rvs(size =1000000), bins=100)
plt.show()
```

#### Fonction de répartition et de densité de probabilité :

```python
X = np.linspace(0,15,5000)
y_pdf = uniform.pdf(X)
y_cdf = uniform.cdf(X)

plt.plot(X, y_pdf, label='pdf')
plt.plot(X, y_cdf, label='cdf')
plt.legend();
```

### Générateur de la loi Normale
#### Histogramme :

```python
mu = 0
sigma = 1
normal = ss.norm(mu, sigma)
plt.hist(normal.rvs(size =1000000), bins=100)
plt.show()
```

#### Fonction de répartition et de densité de probabilité :

```python
X = np.linspace(-5, 5, 5000)

y_pdf = normal.pdf(X)
y_cdf = normal.cdf(X)

plt.plot(X, y_pdf, label='pdf')
plt.plot(X, y_cdf, label='cdf')
plt.legend();
```

### Générateur de la loi Exponentielle
#### Histogramme :

```python
mu = 5
exponential = ss.expon(scale=1./mu)
plt.hist(exponential.rvs(size =1000000), bins=100)
plt.show()
```

#### Fonction de répartition et de densité de probabilité :

```python
X = np.linspace(0, 2, 5000)

y_pdf = exponential.pdf(X)
y_cdf = exponential.cdf(X)

plt.plot(X, y_pdf, label='pdf')
plt.plot(X, y_cdf, label='cdf')
plt.legend();
```

### Générateur de la loi d'Erlang
#### Histogramme :

```python
k = 3
mu = 5
erlang = ss.erlang(a=k, scale=1./mu)
plt.hist(erlang.rvs(size =1000000), bins=100)
plt.show()
```

#### Fonction de répartition et de densité de probabilité :

```python
X = np.linspace(0, 3, 5000)

y_pdf = erlang.pdf(X)
y_cdf = erlang.cdf(X)

plt.plot(X, y_pdf, label='pdf')
plt.plot(X, y_cdf, label='cdf')
plt.legend();
```

### Générateur de la loi de Poisson
#### Histogramme :

```python
#k = 5
mu = 10
poisson = ss.poisson(mu)
plt.hist(poisson.rvs(size =1000000), bins=100)
plt.show()
```

#### Fonction de répartition et de densité de probabilité :

```python
X = np.linspace(0, 20, 5000)

y_cdf = poisson.cdf(X)

plt.plot(X, y_cdf, label='cdf')
plt.legend();
```

## Simuation de file d'attente 
### Modèle M/M/3//FIFO
**Service d’ophtalmologie d’un hôpital :** 
Le service d’ophtalmologie d’un hôpital offre des tests de glaucome gratuits
chaque mardi. Dans ce service travaillent trois ophtalmologues. Le temps
nécessaire pour faire un test de glaucome est exponentiel de moyenne 20
minutes. Les clients arrivent selon un processus de Poisson de taux $\lambda$ = 6
clients/heure et son servis dans l’ordre premier arrivé-premier servi.
1. Calculer le nombre moyen de clients dans la file nf.
2. Calculer le temps de séjour moyen d’un client dans le système ts.
3. Calculer le temps (moyen) d’inoccupation pour chaque médecin

On commence Par quelque import et déclaration de variables servant au monitoring et au paramétrage.

```python
import random, math
import numpy as np
import scipy as sc
import scipy.stats as ss
import simpy as sm
import statistics
from itertools import count


sys_times = []
nbr_queue = []
request_times = []
release_times = []
test_rate = 1. / 20
arrival_rate = 6. / 60
```

Ensuite, on déclare la classe qui représentera la ressource Ophtalmologue.

```python
class Ophtalmo(sm.Resource):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)

    def request(self, *args, **kwargs):
        request_times.append(self._env.now)
        nbr_queue.append(len(self.queue))
        return super().request(*args, **kwargs)

    def release(self, *args, **kwargs):
        release_times.append(self._env.now)
        return super().release(*args, **kwargs)
```

Après cela, nous déclarons le procésus gérant l'Hopitale.

```python
class Hospital(object):
    def __init__(self, env, nbr_ophtalmo):
        self.env = env
        self.ophtalmo = Ophtalmo(env, capacity=nbr_ophtalmo)

    def test_glaucome(self, patient):
        yield self.env.timeout(random.expovariate(test_rate))


def enter_hospital(env, patient, hospital):
    arrival_time = env.now
    with hospital.ophtalmo.request() as request:
        yield request
        yield env.process(hospital.test_glaucome(patient))
    sys_times.append(env.now - arrival_time)


def run_hospital(env, nbr_ophtalmo, nbr_patients):
    hospital = Hospital(env, nbr_ophtalmo)
    patient = 0
    while patient<=nbr_patients:
        yield env.timeout(random.expovariate(arrival_rate))
        patient += 1
        env.process(enter_hospital(env, patient, hospital))
```

Enfin, nous déclarons deux fonctions de calcule des moyennes nf et ts en utilisant les donnée récolé dans *nb_queue* et *sys_times* et nous executons la similation.

```python
def calculate_average_Sys_time(sys_times):
    average_wait = statistics.mean(sys_times)
    # Pretty print the results
    minutes, frac_minutes = divmod(average_wait, 1)
    seconds = frac_minutes * 60
    return round(minutes), round(seconds)


def calculate_average_nbr_nbr_queue(nbr_queue):
    return statistics.mean(nbr_queue)



# Setup
# Run the simulation
env = sm.Environment()
env.process(run_hospital(env, nbr_ophtalmo=3, nbr_patients=100000))
env.run()

# View the results
mins, secs = calculate_average_Sys_time(sys_times)
print("Running simulation...", f"\nThe average system time,ts, is {mins} minutes and {secs} seconds.",
     f"\nThe average number of patient in queue,ns, is {calculate_average_nbr_nbr_queue(nbr_queue)}.")
```

En conclusion, nous pouvant dire que ces résultat sont très similaire au résultats théorique.
